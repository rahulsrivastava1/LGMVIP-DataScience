## Introduction 
In this assignment,i analyse the iris flower dataset , it is a multivariate dataset introduced by Sir Ronald Fisher in 1936 as an example of discriminant analysis.The data set consists of 50 samples from each of three species of Iris (Iris Setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

## Training the Model

```python
from sklearn.model_selection import train_test_split
```

```python
a=iris.drop('species',inplace=False,axis=1)
b=iris['species']
```

```python
a_train,a_test,b_train,b_test=train_test_split(a,b,test_size=0.30)
```

## applying the SVM model to the data.

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(a_train,b_train)
```

```python
pred=svc.predict(a_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(b_test,pred))
```

```python
print(confusion_matrix(b_test,pred))
```

The model has already proved to be excellent, but as an example I will apply GridSearch, which is a methodology used when the data is not so good and if you want to have better results.

```python
from sklearn.model_selection import GridSearchCV 
```

```python
param_grid = {'C' : [0.1, 1, 10, 100, 100], 'gamma' : [1, 0.1, 0.01, 0.001]}
```

```python
grid_cv = GridSearchCV(SVC(), param_grid, refit=True, verbose=2)
```

```python
grid_cv.fit(a_train, b_train)
```

```python
pred = grid_cv.predict(a_test)
```

```python
print(classification_report(b_test, pred))
```

```python
print(confusion_matrix(b_test, pred))
```

From the results above, we see that the GridSearch method was not worth it, as it proved to be the same as the previous result.

So we concluded our project to apply the SVM model to the iris data.
