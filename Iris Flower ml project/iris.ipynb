# Iris Flower Data visualization using python

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
iris=pd.read_csv('IRIS.csv')
```

```python
iris.head()
```

## plotting univariate distributions

```python
sns.distplot(iris['petal_length'])
```

```python
sns.distplot(iris['sepal_width'],kde=False ,rug=True)
```

## Plotting bivariate distributions

```python
sns.jointplot(iris['petal_length'],iris['sepal_width'])
```

```python
sns.jointplot(iris['petal_length'],iris['sepal_width'],kind='hex')
```

## kernel density Estimation

```python
sns.jointplot(iris['petal_length'],iris['sepal_width'],kind='kde')
```

# Visualizing pairwise relationships in a dataset

```python
sns.pairplot(iris[['sepal_length','sepal_width','petal_length','petal_width']])
```

## Plotting with categorical Data

```python
sns.stripplot(iris['sepal_length'],iris['sepal_width'])
```

```python
sns.stripplot(iris['sepal_length'],iris['sepal_width'], jitter=True)
```

```python
sns.swarmplot(iris['sepal_length'],iris['sepal_width'])
```

## Box Plots

```python
sns.boxplot(iris['sepal_length'],iris['sepal_width'], hue=iris['species'])
```

## Bar Plots

```python
sns.barplot(iris['sepal_length'],iris['sepal_width'], hue=iris['species'])
```

```python
sns.countplot(iris['petal_length'])
```

## point plots

```python
sns.pointplot(iris['sepal_length'],iris['sepal_width'], hue=iris['species'])
```

##

```python
sns.factorplot(x="petal_length", 
               y="petal_width", 
               hue="sepal_length", 
               col="sepal_width", 
               data=iris, 
               kind="swarm")
```
